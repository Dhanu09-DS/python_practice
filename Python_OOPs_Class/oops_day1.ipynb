```python
## class car (displya_info, accelerate, brake,)

class Car:
    ## constructor - runs automatically when object is creaetd 
    def __init__(self, brand, model, color):
        self.brand = brand
        self.model = model 
        self.color = color
        self.speed = 0
    
        
    ## Instance method (first task)
    def display_info(self):
        print(f"Car: {self.brand}{self.model}")
        print(f"Color: {self.color}")
        print(f"Current Speed: {self.speed}")
        
    ## task 2 to accelerate 
    def accelerate(self, increase): # 50 
        self.speed += increase
        print(f"{self.brand} is now going at {self.speed} km/h")
        
    ## task 3 brake 
    def brake(self, decrease):      # 30 
        self.speed -= decrease 
        if self.speed < 0:
            self.speed = 0 
        print(f"{self.brand} slowed down to {self.speed} km/h")
        
        
# our class is ready which is blue print for any car you create(object)
```

```python
## class Student(add_subject, add_marks, calculate_avg, display_info)

class Student:
    ## constructor 
    def __init__(self, name, roll_number, grade):
        # Instance attribuets 
        self.name = name 
        self.roll_number = roll_number 
        self.grade = grade 
        self.subjects = []       # empty list to store subjects ["Englis", "maths", "social"]
        self.marks = {}         # Empty dicst to store marks {"maths": 50, "englist": 30}
        
    ## method to add a subject  
    def add_subject(self, subject_name):
        if subject_name not in self.subjects:
            self.subjects.append(subject_name)
            print(f"Subject '{subject_name}' added for {self.name}")
        else:
            print(f"{self.name} is already enrolled in {subject_name}")
            
    ## Method to add marks for a subject 
    def add_marks(self, subject_name, marks):
        if subject_name in self.subjects:
            self.marks[subject_name] = marks 
            print(f"Marks updated: {subject_name} = {marks}")
        else:
            print(f"Error: {subject_name} not found in enrolled subjects")
            
    ## Mehtod to calculate average 
    def calculate_average(self):
        if len(self.marks) == 0:
            return 0 
        total = sum(self.marks.values())
        average = total/len(self.marks)
        return average

    # method to display studnett info 
    def display_info(self):
        print(f"Student Name: {self.name}")
        print(f"Roll number : {self.roll_number}")
        print(f"Grade : {self.grade}")
        print(f"subjects enrolled: {','.join(self.subjects) if self.subjects else "None"}")
        if self.marks:
            print("Marks:")
            for subject, mark in self.marks.items():
                print(f" - {subject}: {mark}")
            print(f"Average: {self.calculate_average():.2f}")
        else:
            print("No marks recorded yet")
            
    # Method to check pass/fail 
    def check_result(self):
        avg = self.calculate_average()
        if avg >= 40:
            print(f" {self.name} has Passed with {avg} %")
        else:
            print(f"{self.name} has failed with {avg}")
```
