# Python Function Practice Questions

1. Write a function that takes a number as input and returns whether it is even or odd.

2. Write a function that takes a number and returns its factorial.

3. Write a function that takes a string and returns the reversed string.

4. Write a function that checks whether a number is prime.

5. Write a function that takes a list of numbers and returns the largest number.

6. Write a function that takes a list and returns the sum of all elements without using the built-in `sum()` function.

7. Write a function that counts the number of vowels in a string.

8. Write a function that returns the Fibonacci sequence up to `n` terms.

9. Write a function that checks whether a string is a palindrome.

10. Write a function that takes two numbers and returns their greatest common divisor (GCD).

11. Write a function that removes duplicates from a list.

12. Write a function that takes a list of numbers and returns a new list containing only even numbers.

13. Write a function that finds the frequency of each character in a string.

14. Write a function that swaps two numbers without using a third variable.

15. Write a function that converts a temperature from Celsius to Fahrenheit.


# 1. Even or Odd

## Logic
- Take a number as input.
- Divide the number by 2.
- Check the remainder using modulus operator `%`.
- If remainder is 0 → number is Even.
- Otherwise → number is Odd.

---

# 2. Factorial of a Number

## Logic
- Take a number `n`.
- Start with a variable `result = 1`.
- Multiply `result` by all numbers from 1 to `n`.
- Return the final value of `result`.
- Special case: factorial of 0 is 1.

**Example:**
5! = 1 × 2 × 3 × 4 × 5

---

# 3. Reverse a String

## Logic

### Method 1 (Loop)
- Take a string.
- Start from the last character.
- Add each character to a new string.
- Return the new reversed string.

### Method 2
- Use slicing with negative step.

---

# 4. Check Prime Number

## Logic
- Prime numbers are greater than 1.
- Check if number is less than or equal to 1 → Not prime.
- Try dividing the number from 2 to `n-1`.
- If divisible by any number → Not prime.
- If not divisible by any → Prime.

---

# 5. Largest Number in List

## Logic
- Assume first element is largest.
- Compare it with each element in the list.
- If a bigger number is found → update largest.
- Return the largest value.

---

# 6. Sum of List (without `sum()`)

## Logic
- Start with `total = 0`.
- Loop through each element.
- Add each element to `total`.
- Return `total`.

---

# 7. Count Vowels

## Logic
- Create a string of vowels (`aeiouAEIOU`).
- Set `count = 0`.
- Loop through each character.
- If character is in vowels → increase `count`.
- Return `count`.

---

# 8. Fibonacci Sequence

## Logic
- First two numbers are 0 and 1.
- Next number = sum of previous two.
- Repeat until required `n` terms.
- Store numbers in a list.
- Return the list.

**Example:**
0, 1, 1, 2, 3, 5, 8…

---

# 9. Palindrome Check

## Logic
- Take a string.
- Reverse the string.
- Compare original string with reversed string.
- If same → Palindrome.
- Else → Not palindrome.

**Example:**
"madam" → same when reversed

---

# 10. GCD (Greatest Common Divisor)

## Logic (Euclidean Method)
- Take two numbers `a` and `b`.
- While `b` is not 0:
  - Replace `a` with `b`.
  - Replace `b` with remainder of `a % b`.
- When `b` becomes 0 → `a` is GCD.

---

# 11. Remove Duplicates from List

## Logic
- Create empty list.
- Loop through original list.
- If element is not already in new list → add it.
- Return new list.

---

# 12. Get Even Numbers from List

## Logic
- Create empty list.
- Loop through original list.
- If number `% 2 == 0` → add to new list.
- Return new list.

---

# 13. Character Frequency in String

## Logic
- Create empty dictionary.
- Loop through each character.
- If character already in dictionary:
  - Increase its count.
- Else:
  - Add character with count 1.
- Return dictionary.

---

# 14. Swap Two Numbers (Without Third Variable)

## Logic

### Method 1 (Python way)
`a, b = b, a`

### Method 2 (Math way)
- `a = a + b`
- `b = a - b`
- `a = a - b`

---

# 15. Celsius to Fahrenheit

## Logic
- Take temperature in Celsius.
- Use formula:
  `Fahrenheit = (Celsius × 9/5) + 32`
- Return the result.
